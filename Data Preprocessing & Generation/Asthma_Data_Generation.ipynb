```python
import numpy as np
import pandas as pd
from datetime import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedShuffleSplit, train_test_split
```

```python
pd.set_option("display.max_rows", 11)
```

# GSE 8052

```python
asthma_data = "8052"
```

```python
asthma_gse_8052_df = pd.read_csv(f"KEGG_Based_Genes_GSE_{asthma_data}_Dataset.csv")
```

```python
asthma_gse_8052_df
```

```python
data = asthma_gse_8052_df.iloc[:, 2:]
```

```python
data
```

```python
label = asthma_gse_8052_df.iloc[:, :2]
```

```python
label
```

```python
experiment = 1
for train, test in StratifiedShuffleSplit(n_splits = 10, test_size = 0.2).split(data, label):
    trainData, validData, trainLabel, validLabel = train_test_split(data.loc[train], label.loc[train], test_size = 0.2, shuffle = True, stratify = label.loc[train])
    testData = data.loc[test]
    testLabel = label.loc[test]
    print("#######################  %d experiment  #######################\n" % experiment)
    ##################################################################################################
    data_scaler = StandardScaler()
    data_scaler.fit(trainData)
    scaled_data_based_asthma_8052_train_data = data_scaler.transform(trainData)
    scaled_data_based_asthma_8052_valid_data = data_scaler.transform(validData)
    scaled_data_based_asthma_8052_test_data = data_scaler.transform(testData)
    ### entire data
    scaled_asthma_8052_train_df = pd.DataFrame(scaled_data_based_asthma_8052_train_data)
    scaled_asthma_8052_valid_df = pd.DataFrame(scaled_data_based_asthma_8052_valid_data)
    scaled_asthma_8052_test_df = pd.DataFrame(scaled_data_based_asthma_8052_test_data)
    ### reindex label
    reindexed_asthma_8052_train_label = trainLabel.reset_index(drop = True)
    reindexed_asthma_8052_valid_label = validLabel.reset_index(drop = True)
    reindexed_asthma_8052_test_label = testLabel.reset_index(drop = True)
    ##################################################################################################
    ### concat entire data with sex information
    scaled_asthma_8052_train_with_sex_df = pd.concat([scaled_asthma_8052_train_df, reindexed_asthma_8052_train_label["Gender"]], axis = 1)
    scaled_asthma_8052_valid_with_sex_df = pd.concat([scaled_asthma_8052_valid_df, reindexed_asthma_8052_valid_label["Gender"]], axis = 1)
    scaled_asthma_8052_test_with_sex_df = pd.concat([scaled_asthma_8052_test_df, reindexed_asthma_8052_test_label["Gender"]], axis = 1)
    ### entire label
    asthma_8052_train_label = reindexed_asthma_8052_train_label["Label"]
    asthma_8052_valid_label = reindexed_asthma_8052_valid_label["Label"]
    asthma_8052_test_label = reindexed_asthma_8052_test_label["Label"]
    ##################################################################################################
    scaled_asthma_8052_train_with_sex_df.to_csv(f"[GSE{asthma_data}]Entire_Normed_Train_Data_{experiment}.csv", index = False, header = True)
    scaled_asthma_8052_valid_with_sex_df.to_csv(f"[GSE{asthma_data}]Entire_Normed_Valid_Data_{experiment}.csv", index = False, header = True)
    scaled_asthma_8052_test_with_sex_df.to_csv(f"[GSE{asthma_data}]Entire_Normed_Test_Data_{experiment}.csv", index = False, header = True)
    asthma_8052_train_label.to_csv(f"[GSE{asthma_data}]Entire_Normed_Train_Label_{experiment}.csv", index = False, header = True)
    asthma_8052_valid_label.to_csv(f"[GSE{asthma_data}]Entire_Normed_Valid_Label_{experiment}.csv", index = False, header = True)
    asthma_8052_test_label.to_csv(f"[GSE{asthma_data}]Entire_Normed_Test_Label_{experiment}.csv", index = False, header = True)
    
    experiment += 1
```
